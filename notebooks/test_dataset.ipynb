# Station test

```python
X_station_test = pd.read_csv("../../Hackathon/Test/X_station_test.csv") #, index_col=["Id"])
X_station_test.head()
```

```python
X_station_test[X_station_test['Id']=='14296001_52_0']
#X_station_test[X_station_test['Id']=='86137003_256_0']
```

```python
print(len(X_station_test[X_station_test['precip'].isnull()].index), "NaN rows")
print(100*len(X_station_test[X_station_test['precip'].notnull()].index)/len(X_station_test.index), "% of measured values without NaN")
```

```python
df_station = X_station_test[X_station_test['precip'].notnull()]
```

```python
def station(row):
    my_str = row['Id']
    return my_str.split('_')[0]

def day(row):
    my_str = row['Id']
    return my_str.split('_')[1]

def station_day(row):
    my_str = row['Id']
    return my_str.split('_')[0]+"_"+my_str.split('_')[1]
```

```python
df_station['station_day'] = df_station.apply(lambda row: station_day(row), axis = 1)
```

```python
df_station.head()
```

```python
def get_station_day(station, day):
    station_day_str = str(int(station)) + "_" + str(int(day))
    df_station_day = df_station[df_station['station_day']==station_day_str]
    return df_station_day

get_station_day(14296001,52)
```

```python
measured_precips = pd.DataFrame(df_station.groupby(by="station_day")['precip'].sum())
```

```python
measured_precips = measured_precips.reset_index()
```

```python
measured_precips[measured_precips['station_day']=='14296001_52']
```

# Merge

```python
df_merge = pd.merge(predicted_precips, measured_precips, how='outer', on='station_day')
df_merge = df_merge.reset_index()
df_merge = df_merge.rename({'station_day': 'Id', 'precip': 'measure', 'tp' : 'prediction_arome'}, axis='columns')
df_merge
```
